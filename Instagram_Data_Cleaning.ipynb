### **Outliers**

```python
# checking outliers for key_metrics numerical columns
outlier_cols = ['likes_given_per_day','comments_written_per_day', 'dms_sent_per_week', 'posts_created_per_week', 'time_on_feed_per_day', 'time_on_reels_per_day',
                'time_on_explore_per_day', 'time_on_messages_per_day', 'dms_received_per_week', 'ads_clicked_per_day', 'ads_viewed_per_day']

outlier_report = []
for col in outlier_cols:
    q1 = instagram_usage[col].quantile(0.25)
    q3 = instagram_usage[col].quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - (1.5 * iqr)
    upper_bound = q3 + (1.5 * iqr)

    mask = ((instagram_usage[col] < lower_bound) | (instagram_usage[col] > upper_bound))
    outliers = instagram_usage.loc[mask,['user_id', col]]

    print(f"\nOutliers count in {col}: {len(outliers)}")
    print(f"Outliers percentage in {col}: {round((len(outliers) * 100 / len(instagram_usage)),2)}")

    print(outliers.head())

```

#### Outliers in main metric columsn are in less percentage. So capping the values to avoid the distortion in the analysis.

#### **Handling Outliers**

```python
# Capping outliers using 1st and 99th percentiles

for col in outlier_cols:

    lower_limit = instagram_usage[col].quantile(0.01)

    upper_limit = instagram_usage[col].quantile(0.99)

    instagram_usage[col] = instagram_usage[col].clip(
        lower=lower_limit,
        upper=upper_limit
    )

print("Capping completed - Outliers Handled")
```
