```python
# importing all necessary libraries
import gdown as gd
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

### **Importing user metrics Cleaned Dataset from analysis in pg admin**

```python
# importing "user_metrics" dataset
file_id = "1cwBDZv6tLL7B-8B-XutvlbuVhX6W7RyO"
download_url = f"https://docs.google.com/uc?id={file_id}"

output_file = "user_metrics.csv"

gd.download(download_url, output_file, quiet=False)
user_metrics = pd.read_csv("user_metrics.csv")
```

```python
user_metrics.shape
```

### **Correlation Analysis**

#### **1. Lifestyle Factors vs Engagement Quality Efficiency (EQE)**

```python
lifestyle=user_metrics[['sleep_hours_per_night',
                        'exercise_hours_per_week',
                        'perceived_stress_score',
                        'eqe']]

lifestyle_corr=lifestyle.corr()

plt.figure(figsize=(6,3))

sns.heatmap(lifestyle_corr,
            annot=True,
            cmap='coolwarm')

plt.title('Lifestyle Factors vs EQE')

plt.show()
```

#### **2. Platform Behavior vs Engagement Quality Efficiency (EQE)**

```python
platform=user_metrics[['sessions_per_day',
                       'daily_active_minutes_instagram',
                       'average_session_length_minutes',
                       'time_on_feed_per_day',
                       'time_on_reels_per_day',
                       'eqe']]

platform_corr=platform.corr()

plt.figure(figsize=(8,4))

sns.heatmap(platform_corr,
            annot=True,
            cmap='coolwarm')

plt.title('Platform Behavior vs EQE')

plt.show()
```

#### **3. Business Metric Validation (ME, Attention & EQE)**

```python
business=user_metrics[['me',
                       'attention',
                       'eqe']]

business_corr=business.corr()

plt.figure(figsize=(5,2))

sns.heatmap(business_corr,
            annot=True,
            cmap='coolwarm')

plt.title('Business Metric Validation')

plt.show()
```

**Observations:**

• Stress shows a moderate negative relationship with EQE.

• Excessive platform usage behaviors are negatively associated with EQE.

• Meaningful Engagement (ME) positively influences EQE, while excessive attention reduces it.
